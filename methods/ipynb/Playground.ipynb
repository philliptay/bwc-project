### Chunk Video Names

```python
import os
import numpy as np
```

```python
vid_dir = '/n/fs/vai-bwc-scr/videos/'
vid_names = os.listdir(vid_dir)
```

```python
def chunk_list(l, num_chunks):
    cp = np.linspace(0, len(l), num_chunks+1, dtype=int)
    cl = []
    for i in range(len(cp)-1):
        cl.append(l[cp[i]:cp[i+1]])
    return cl
```

```python
cl = chunk_list(vid_names, 10)
```

```python
chunk_dir = '/n/fs/vai-bwc-scr/methods/vid_name_chunks'
chunk_name = 'chunk_{}.txt'
for i, l in enumerate(cl):
    with open(os.path.join(chunk_dir, chunk_name.format(i)), 'w') as f:
        for vid_name in l:
            f.write(vid_name + '\n')
```
