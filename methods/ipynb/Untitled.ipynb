```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import pdb

def sample_middle(frame_paths, num_frames):
    if len(frame_paths) <= num_frames:
        return frame_paths
    middle_idx = len(frame_paths) // 2
    return frame_paths[middle_idx-num_frames//2:middle_idx+num_frames//2]
```

```python
data_dir = '/n/fs/visualai-scr/Data/BWC/frames/'
id_title_filepath = '/n/fs/vai-bwc-scr/annotations/URLs/ids_and_titles.txt'
third_filepath = '/n/fs/vai-bwc-scr/annotations/URLs/3rdparty_urls.txt'
```

```python
title_to_id = {}
id_to_title = {}
with open(id_title_filepath, 'r') as f:
    for line in f:
        line = line.strip()
        vid_id = line[0:9]
        vid_name = line[10:]
        vid_name = vid_name.replace(' ', '_')
        vid_name = ''.join(vid_name.split())
        title_to_id[vid_name] = vid_id
        id_to_title[vid_id] = vid_name
```

```python
third_ids = []
with open(third_filepath, 'r') as f:
    for line in f:
        vid_id = line.strip().split('/')[-1]
        third_ids.append(vid_id)
```

```python
frame_dirs = []
for third_id in third_ids:
    frame_dir = os.path.join(data_dir, id_to_title[third_id])
    if os.path.isdir(frame_dir):
        frame_dirs.append(frame_dir)
    else:
        print('Vid name {} was never downloaded.'.format(id_to_title[third_id]))
```

```python
diffs = []
num_frames = 20
idx = 1
for frame_dir in frame_dirs:
    print('Looking at idx {} video {}'.format(str(idx).zfill(3), frame_dir.split('/')[-1]))
    frame_paths = [os.path.join(frame_dir, '{}.jpg'.format(x)) for x in range(1, len(os.listdir(frame_dir))+1)]
    frame_paths = sample_middle(frame_paths, num_frames)
    prev_img_np = None
    diff = []
    for frame_path in frame_paths:
        with Image.open(frame_path) as img:
            img_np = np.asarray(img, dtype=float)
        if prev_img_np is not None:
            diff.append(np.mean(np.abs(prev_img_np - img_np)))
        prev_img_np = img_np
    diffs.append(np.mean(np.array(diff)))
    idx += 1
```

```python
plt.figure(figsize=(20,5))
plt.hist(diffs, bins=40)
```

```python
for i, frame_dir in enumerate(frame_dirs):
    vid_name = frame_dir.split('/')[-1].replace('_', ' ')
    diff = diffs[i]
    print('{}: {:.2f}'.format(vid_name, diff))
```

```python
prefix = 'https://vimeo.com/'
vid_ids = []
for i, frame_dir in enumerate(frame_dirs):
    diff = diffs[i]
    if diff >= 10.0:
        vid_ids.append(title_to_id[frame_dir.split('/')[-1]])
```

```python
with open('filtered_3rd_party_urls.txt', 'w') as f:
    for vid_id in vid_ids:
        f.write('{}{}\n'.format(prefix, str(vid_id)))
```
